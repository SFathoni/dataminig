```python
import pandas as pd
from scipy import stats
import statistics
df=pd.read_excel("data_M5.xlsx", usecols=[1,2,3,4,5])

df

for i in df:
    print('---%s---'%i) #to print title
    Q1=df[i].quantile(0.25)
    print('Q1 =',Q1)
    Q2=df[i].quantile(0.5) 
    print('Q2 =',Q2)
    Q3=df[i].quantile(0.75) 
    print('Q3 =',Q3)
    Rata=df[i].mean()
    print('Mean =',Rata)
    Me=df[i].median()
    print('Median =',Me)
    Std_dev=statistics.stdev(df[i])
    print('Std Dev =',Std_dev)
    Skewness=df[i].skew()
    print('Skewness =',Skewness,'\n')
```
